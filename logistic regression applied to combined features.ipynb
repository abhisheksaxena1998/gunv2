```python
import pandas as pd
import pandas as pd
import numpy as np

#df1=pd.read_csv("cleaned_reddit_alphabeta.csv")
df1=pd.read_csv("cleaned_reddit_alphabetav5.csv")
#df1=pd.read_csv("https://raw.githubusercontent.com/abhisheksaxena1998/detector-redditv3/master/reddit_cleaned_lower_combined.csv")
df1.dropna(inplace=True)
df1.columns=['index','combined','flair']  
print (len (df1))
df1
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df1['combined'], 
                                                    df1['flair'], 
                                                    random_state=0)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Fit the CountVectorizer to the training data
vect = CountVectorizer().fit(X_train)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier


X_train_vectorized = vect.transform(X_train)

X_train_vectorized.todense()
from sklearn.metrics import roc_auc_score
from sklearn import preprocessing

def multiclass_roc_auc_score(y_test, y_pred, average="macro"):
    lb = preprocessing.LabelBinarizer()
    lb.fit(y_test)
    y_test = lb.transform(y_test)
    y_pred = lb.transform(y_pred)
    return roc_auc_score(y_test, y_pred, average=average)
    
model = LogisticRegression(solver='lbfgs', multi_class='auto',max_iter=1000)
    
model.fit(X_train_vectorized, y_train)

predictions = model.predict(vect.transform(X_test))

print('AUC: ', multiclass_roc_auc_score(y_test, predictions))
```
